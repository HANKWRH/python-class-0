# Data visualization with [Matplotlib](https://matplotlib.org/)

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#color https://www.w3schools.com/colors/colors_picker.asp
height_data = [100, 200, 150, 50, 70, 90]
x_labels = ['TW', 'USA', 'JAP', 'KOR', 'UK', '']
x_positions = range(len(x_labels))


plt.title('Apple user', fontdict={'fontsize' : 50})
plt.xticks(x_positions, x_labels)
plot = plt.bar(x_positions, height_data, bottom=0, color='#00ff99')
```

```python
def create_bar(title, x_labels, color, height, fontsize):
    x_positions = range(len(x_labels))

    plt.title(title, fontdict={'fontsize' : fontsize})
    plt.xticks(x_positions, x_labels)
    plt.bar(x_positions, height, bottom=0, color=color)
  
```

```python
height_data = [100, 200, 150, 50, 70, 990]
x_labels = ['TW', 'USA', 'JAP', 'KOR', 'UK', '']
color='#00ff99'
create_bar("title",x_labels,color,height_data,15)
```

```python
height_data = [9, 5, 6, 11, 4]
x_labels = ['soccer', 'baseball', 'basketball', 'dodgeball', 'volleyball']
color='#3399ff'
create_bar("The sport people like in class",x_labels,color,height_data,15)
```

```python
height_data = [100, 200, 150, 50, 70, 90]
x_labels = ['TW', 'USA', 'JAP', 'KOR', 'UK', '']
x_positions = range(len(x_labels))


plt.title('Apple user', fontdict={'fontsize' : 50})
plt.xticks(x_positions, x_labels)
plt.plot(x_positions, height_data, color='#00ff99')
```

```python
def create_lines(title, x_labels, color, height, fontsize):
    x_positions = range(len(x_labels))

    plt.title(title, fontdict={'fontsize' : fontsize})
    plt.xticks(x_positions, x_labels)
    plt.plot(x_positions, height, color=color)
```

```python
height_data = [9, 5, 6, 11, 4]
x_labels = ['soccer', 'baseball', 'basketball', 'dodgeball', 'volleyball']
color='#3399ff'
create_lines("The sport people like in class",x_labels,color,height_data,15)
```

```python
from math import sin, cos
```

```python
x_values = [i * 0.1 for i in range(-100, 0)] + [i * 0.1 for i in range(0,100)]
y_values = [sin(x) for x in x_values]
plt.plot(x_values, y_values)
```

```python
def plot_func(x_values, func):
    y_values = [func(x) for x in x_values]
    plt.plot(x_values, y_values)

plot_func(x_values, cos)
```

```python
def my_func(x):
    return sin(x*x)
plot_func(x_values, my_func)
```
